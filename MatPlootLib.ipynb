##Importando os pacotes

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##Grafico de linhas

```python
np.random.seed(7)

#Criando vetor com numeros aleatorios entre 1 e 1500 com 10 amostras

y = np.random .randint(low=1, high=1500, size=10)
```

```python
y
```

```python
plt.plot(y)
```

###Configurando o plot

```python
# insere a primeira linha
plt.plot(y, color='#749187', marker='o', ms=5, mec='k', markerfacecolor='w', ls='-.')

#insere a segunda linha
plt.plot(y*2, marker='+', color='k', ms=5)

#rotulos
plt.xlabel('Eixo X', color='red', size=12)
plt.ylabel('Eixo Y', color='green', size=12)
plt.title('Treinando Plot', loc='center')

#gridlines
plt.grid(axis='y', color='gray', linestyle='--', linewidth=1, alpha=0.8)

#mostrar o grafico
plt.show()
```

###Alinhando 3 plots


```python
#Criando 3 base de dados aleatorias
np.random.seed(6)
x = np.arange(1,11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

#Configurando a figura
plt.figure(figsize =(20,5))
plt.suptitle('3 Plots', fontsize = 15)

#Carrengando cada um dos plots
plt.subplot(1,3,1)
plt.plot(x, y1, color = 'k')
plt.title('1 Plot', pad=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')


plt.subplot(1,3,2)
plt.plot(x, y2, color = 'r')
plt.title('2 Plot', pad=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')


plt.subplot(1,3,3)
plt.plot(x, y3, color = 'g')
plt.title('3 Plot', pad=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
```

##Grafico Scatter Plot

```python
#Recebendo os dados aleatoriamente

linha = np.random.normal(10, 0.5, 100)
coluna = np.random.normal(0, 20, 100)
```

```python
#Montando o plot

fig = plt.figure()
ax = plt.axes()
ax.scatter(linha, coluna, marker='o', color='red', label='data 1', alpha = 0.5)
ax.scatter(linha*0.5, coluna*0.5, marker='v', color='green', label='data 2', alpha = 0.7)
ax.legend();
```

##Teste com dados Reais

```python
import plotly.express as px
```

```python
#importando um dataset
px.data.gapminder()
```

```python
#DataFrame filtrando o Brasil
df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
#Criando o plot
plt.plot(df.index, df['gdpPercap'])
plt.title('PIB per capita do Brasil')
plt.ylabel('PIB per capita')
plt.xlabel('Tempo')
```

```python
df.head()
```

```python
title = 'Relação entre a expectativa de vida e a renda per capita no Brasil'

plt.figure(figsize = (12,4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap='viridis')
plt.ylabel('Renda per capita')
plt.title(title, loc='left')
plt.show()
```

```python
#Grafico de barras

plt.bar(x = df.index, height = df['pop'], color='green')
plt.title('População brasileira')
```

##Criando funcões para analise

```python
def filtrar_continente(continente):
  df = px.data.gapminder()
  df = df[df['continent'] == continente]
  return df
```

```python
def filtrar_pais(pais, variavel):
  df = px.data.gapminder()
  df= df[df['country'] == pais][variavel]
  return df
```

```python
americas = filtrar_continente('Americas')
paises = americas['country'].unique()
plt.figure(figsize = (12,8)) 

for pais in paises: 
  plt.scatter(filtrar_pais(pais = pais, variavel = 'lifeExp'),
              filtrar_pais(pais = pais, variavel = 'gdpPercap'))
  
plt.legend(labels = paises, loc = 'best')
plt.title('Relação entre renda per capita e expectativa de vida', loc = 'left')
plt.xlabel('Expectativa de vida')
plt.ylabel('Renda per capita')
```
